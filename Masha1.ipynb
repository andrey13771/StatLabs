```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas_profiling
from statsmodels.graphics.gofplots import qqplot
import statsmodels.api as sm
from scipy import stats
```

```python
df = pd.read_excel('~/Downloads/distrib.xls')
```

```python
la = df['La']
la = la.dropna()
sns.distplot(la)
```

```python
la = la - 1
sns.distplot(la)
```

```python
qqplot(la, line='s', loc=0, scale=0.8)
```

```python
qqplot(la, 'laplace', line='s', loc=0, scale=0.8)
```

```python
stats.normaltest(la)
```

```python
stats.shapiro(la)
```

відкидаємо гіпотезу про нормальність розподілу

```python
stats.kstest(la, 'laplace')
```

отримали p-value > 0.05, отже можемо припускати що дані мають розподіл лапласа
